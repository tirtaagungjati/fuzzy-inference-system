# Fungsi Segitiga

```python
def fungsi_segitiga(x,a,b,c):
    def kalimat():
        print("Berikut Gambar Kurva dari Fungsi yang Dibuat")
        plt.plot([a,b,c],[0,1,0])
        plt.ylabel('Fire Strength')
    if x == "x":
        kalimat()
    else:
        fire_strength = max(min((x-a)/(b-a),(c-x)/(c-b)),0)
        plt.plot([x], [fire_strength], 'ro')
        print("Hasil Fire strength dari",x,"adalah",fire_strength)
        kalimat()
```

```python
fungsi_segitiga("x",2,5,8)
```

```python
fungsi_segitiga(5,2,5,8)
```
